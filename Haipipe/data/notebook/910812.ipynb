#                                 All Necessary libraries of scikit-learn

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

# Reading CSV Data set 

```python
df=pd.read_csv("../input/security.csv")
df.shape
```

# Converting string Values of Below Columns into Numerical values using Pandas Library

```python
#Using Pandas get dummies function to convert the string columns into numerics

new_df=pd.get_dummies(df, columns=['DOMAIN_NAME','URL','WHOIS_COUNTRY','WHOIS_STATE_CITY','SERVER'])
```

# Method to increase the display size in jyputer Notebook

```python
pd.options.display.max_seq_items=100000
pd.options.display.max_seq_items
```

# Converting the Label #Benigna 0 and Maligna 1 into category data

```python
new_df['TIPO']= new_df['TIPO'].astype('category')
```

# Saving the categorical value into new Column called Target

```python
new_df['Target']=new_df['TIPO'].cat.codes
new_df.head()
```

# Creating Duplicate Data frame 

```python
target_df=new_df
target_df.columns
```

# Droping the columns that gives less information about the data set

```python
new_df=new_df.drop(['TIPO','CHARSET','CACHE_CONTROL','Target'], axis=1)
```

# Storing all the columns of New data frame into variable Fetaure_Cols to make Feature matrix

```python
feature_cols=new_df.columns
```

```python
x=new_df[feature_cols]
x.shape
```

# Target matrix

```python
#Benigna       0
#Maligna       1
y=target_df['Target']
y.shape
```

# splitting the data set into two training and Testing Data.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=2)

# print the size of the traning set:
print(x_train.shape)
print(y_train.shape)

# print the size of the testing set:
print(x_test.shape)
print(y_test.shape)
```

# Creating models using different Machine Learning Algorithm and fit them with training data

```python
dec_tree=DecisionTreeClassifier() 
log_reg=LogisticRegression()
k=5
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
log_reg.fit(x_train,y_train)
dec_tree.fit(x_train,y_train)
knn.fit(x_train,y_train)
```

# Testing the fit model on testing Data 

```python
y_perdict_regression =log_reg.predict(x_test)
y_perdict_tree =dec_tree.predict(x_test)
y_predict_knn= knn.predict(x_test)
```

# Accuracy Score of all the trained model

```python
score_regression=accuracy_score(y_test,y_perdict_regression)
score_tree=accuracy_score(y_test, y_perdict_tree)
score_knn=accuracy_score(y_test, y_predict_knn)

print('Logistic Regression',score_regression)
print('Decision Tree',score_tree)
print('KNN',score_knn)
```
