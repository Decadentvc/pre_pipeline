```python
data = pd.read_csv("../input/Social_Network_Ads.csv")
```

```python
data.head()
```

```python
#encode gender in numeric data
le = preprocessing.LabelEncoder()
data['Gender'] = le.fit_transform(data['Gender'])
```

```python
#seperating features and labels
X = data.iloc[:, [2, 3]].values
y = data.iloc[:, 4].values
```

**Run Either of both for seperating features and labels (just for explaination purpose)**

```python
X_features = ['Gender','Age','EstimatedSalary']
y_features = ['Purchased']
X = data[X_features]
y = data[y_features]
```

```python
#plotting the distribution of feature AGE.
sns.distplot(X['Age']);
```

```python
sns.distplot(X['EstimatedSalary'], bins=10, kde=False, rug=True);
```

**Spliting the 80% data for training and 20% for testing**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

**Fitting Logistic Regression to the Training set**

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

**Prediction**

```python
y_pred = classifier.predict(X_test)
```

**Evaluation**

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_pred, y_test)
```
