```python
#concatenate dummies to dataframe
df = pd.concat([df,cp,restecg,slope,ca,thal],axis=1)
```

```python
#logistic regression model
```

```python
#import
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
#scale data
StSc = StandardScaler()
```

```python
#variables to be scaled
scaled_num_col = ['age','trestbps','chol','thalach','oldpeak']
```

```python
#scale selected variables
df[scaled_num_col] = StSc.fit_transform(df[scaled_num_col])
```

```python
#examine scaled data
df.describe()
```

```python
#independent variables
X = df.drop('target',axis=1)
```

```python
#dependent variable
y = df['target']
```

```python
#train/test split of the data
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
#import
from sklearn.linear_model import LogisticRegression
```

```python
#logit model
logmodel = LogisticRegression()
```

```python
#model fit
logmodel.fit(X_train,y_train)
```

```python
#call predictions
predictions=logmodel.predict(X_test)
```

```python
#classification report
print (classification_report(y_test,predictions))
```

```python
#compare distribution in data set - looks good
df['target'].value_counts(normalize=True) * 100
```

```python
#print confusion matrix - looks good
confusion_matrix (y_test,predictions)
```

```python
#decision tree model
```

```python
#import
from sklearn.tree import DecisionTreeClassifier
```

```python
#decision tree
dtree = DecisionTreeClassifier()
```

```python
#model fit
dtree.fit(X_train,y_train)
```

```python
#call predictions
predictions = dtree.predict(X_test)
```

```python
#print classification report
print(classification_report(y_test,predictions))
```

```python
#print confusion matrix
print(confusion_matrix(y_test,predictions))
```

```python
#random forest model
```

```python
#import
from sklearn.ensemble import RandomForestClassifier
```

```python
#random forest model
rfc = RandomForestClassifier(n_estimators=300)
```

```python
#model fit
rfc.fit(X_train,y_train)
```

```python
#call predictions
predictions = rfc.predict(X_test)
```

```python
#print classification report
#notice an improvement compared to decesion tree model
print(classification_report(y_test,predictions))
```

```python
#print confusion matrix
print(confusion_matrix(y_test,predictions))
```
