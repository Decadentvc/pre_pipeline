**BOXPLOTS**

Boxplot veri dalmn grntlemenin  bir yoludur. Size aykr deerleriniz ve deerlerinin ne olduu hakknda bilgi verebilir. Ayrca verilerinizin simetrik olup olmadn, verilerinizin ne kadar sk bir ekilde gruplandn ve verilerinizin eri olup olmadn ve nasl eildiini size syleyebilir.

Kutu grafikleri, be say zetine (minimum, ilk eyrek (Q1), median, nc eyrek (Q3) ve maksimum) dayal olarak veri dalmn grntlemenin standart bir yoludur.


* Medyan (Q2 / 50. Yzdelik) : veri kmesinin orta deeri.

* lk eyrek (Q1 / 25. Yzde): en kk say (minimum deil) ve veri kmesinin ortancas arasndaki orta say.

* nc eyrek (Q3 / 75. Yzdelik): veri kmesinin ortanca deeri ile en yksek deeri (maksimum deil) arasndaki orta deer.

* eyrekler aras aralk (IQR): 25. ila 75. persentil.

* whiskers (mavi renkle gsterilir)

* Outliers(aykr deerler) : yeil daireler olarak gsterilir.
* Maksimum: Q3 + 1,5 * IQR
* Minimum: Q1 -1,5 * IQR
* Outlier'ler verinin %7 sini oluturur..

**K-NEAREST NEIGHBORS (KNN)**

Algoritmann almasnda bir K deeri belirlenir. Bu K deerinin anlam baklacak eleman saysdr. Bir deer geldiinde en yakn K kadar eleman alnarak gelen deer arasndaki uzaklk hesaplanr. Uzaklk hesaplandktan sonra sralanr ve gelen deer uygun olan snfa atanr.


```python
from sklearn.model_selection import train_test_split # Datamz train ve test olarak blyoruz.
x,y = data.loc[:,data.columns != 'target'], data.loc[:,'target']
x_train, x_test, y_train, y_test  =train_test_split(x,y, test_size =0.3 , random_state = 42)

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5) # n_neighbors : K deeridir. Baklacak eleman saysdr.
knn.fit(x_train,y_train)
prediction = knn.predict(x_test)

print('K=5 iin doruluk : ',knn.score(x_test,y_test)) 
```

Peki en uygun K deerini nasl bulabiliriz ?

```python
# k'y 1'dan 25'e kadar seiyoruz ve bizim iin en uygun deeri bulalm.

aralk = np.arange(1,25)
train_dogruluk =[]
test_dogruluk = []

for i ,k in enumerate(aralk):
    knn = KNeighborsClassifier(n_neighbors=k)
    # knn ile fit ediyoruz.
    knn.fit(x_train,y_train)
    #train doruluk
    train_dogruluk.append(knn.score(x_train, y_train))
    # test doruluk
    test_dogruluk.append(knn.score(x_test, y_test))
    
# imdi doruluk grafiini izdireceiz



plt.figure(figsize=[13,8])
plt.plot(aralk, test_dogruluk, label = 'Test Doruluu')
plt.plot(aralk, train_dogruluk, label = 'Training Doruluu')
plt.legend()
plt.title('-value VS Doruluk')
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.xticks(aralk)
plt.savefig('graph.png')
plt.show()
print("En iyi doruluk {} with K = {}".format(np.max(test_dogruluk),1+test_dogruluk.index(np.max(test_dogruluk))))

    
    
```

**SUPPORT VECTOR MACHNE (SWM) CLASSFCATON**

Snflandrma (Classification) konusunda kullanlan olduka etkili ve basit yntemlerden birisidir. Snflandrma iin bir dzlemde bulunan iki grup arasnda bir snr izilerek iki grubu ayrmak mmkndr. Bu snrn izilecei yer ise iki grubun da yelerine en uzak olan yer olmaldr. te SVM bu snrn nasl izileceini belirler.

Bu ilemin yaplmas iin iki gruba da yakn ve birbirine paralel iki snr izgisi izilir ve bu snr izgileri birbirine yaklatrlarak ortak snr izgisi retilir.

```python
#SWM
from sklearn.svm import SVC

svc = SVC(random_state = 42)

svc.fit(x_train,y_train)

print("SWM modelinin dorulu {}".format(svc.score(x_test,y_test)))

result = svc.predict(x_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,result)
print(cm)
```


**NAVE BAYES CLASSFCATON **

Naive Bayes snflandrcsnn temeli Bayes teoremine dayanr. lazy ( tembel ) bir renme algoritmasdr ayn zamanda dengesiz veri kmelerinde de alabilir. Algoritmann alma ekli bir eleman iin her durumun olasln hesaplar ve olaslk deeri en yksek olana gre snflandrr. Az bir eitim verisiyle ok baarl iler kartabilir. Test kmesindeki bir deerin eitim kmesinde gzlemlenemeyen bir deeri varsa olaslk deeri olarak 0 verir yani tahmin yapamaz. Bu durum genellikle Zero Frequency ( Sfr Frekans ) adyla bilinir.
```python
# NAVE BAYES
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(x_train,y_train)

print("Naive Bayes modelinin doruluu {}".format(nb.score(x_test,y_test)))

# Confusion Matrix
result = svc.predict(x_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,result)
print(cm)
```

**Decision Tree**
 
Decision Tree'nin amac veri zelliklerinden basit kurallar karp bu kurallar renerek bir deikenin deerini tahmin eden modeli oluturmaktr.

```python
#Decision Tree

from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(random_state=42)

dtc.fit(x_train,y_train)

result = dtc.predict(x_test)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test,result)
print(cm)

print("Decision Tree modeli doruluk oran {}".format(dtc.score(x_test,y_test)))
```
